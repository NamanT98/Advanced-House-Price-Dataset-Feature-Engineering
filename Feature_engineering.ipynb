# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("train.csv")
```

```python
pd.pandas.set_option("display.max_columns",None)
```

# Missing values

## Categorical

```python
null_cat=[i for i in dataset.columns if dataset[i].isnull().sum()>1 and dataset[i].dtype=="O"]
len(null_cat)
```

```python
for i in null_cat:
  print(i," ",dataset[i].isnull().sum())
```

```python
def filling_missing_val(dataset,features):
  data=dataset.copy()
  data[features]=data[features].fillna("Missing")
  return data

dataset=filling_missing_val(dataset,null_cat)
dataset.isnull().sum()
```

```python
dataset[null_cat].isnull().sum()
```

```python
dataset.head()
```

## Numerical 

```python
null_num=[i for i in dataset.columns if dataset[i].isnull().sum()>1 and dataset[i].dtype!="O"]
len(null_num)
```

```python
dataset[null_num].head()
```

```python
# we are making new columns for knowing which values were replaced in numerical features.
```

```python
for i in null_num:
  median=dataset[i].median()
  # we are replacing with median because there are outliers and large values can change mean drastically
  dataset[i+"_nan"]=np.where(dataset[i].isnull(),1,0)
  dataset[i].fillna(median,inplace=True)

dataset[null_num].isnull().sum()
```

# Temporal variable (Date,time,etc)

```python
year=[i for i in dataset.columns if "Year" in i or "Yr" in i and i!="YrSold" and "_nan" not in i]
year
```

```python
for i in year:
  dataset[i]=dataset["YrSold"]-dataset[i]
dataset[year].head()
```

# Numerical values


```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea',"SalePrice"]
for i in num_features:
  dataset[i]=np.log(dataset[i])
```

```python
dataset.head()
```

# Rare categorical data

```python
categorical=[i for i in dataset.columns if dataset[i].dtype=="O"]
len(categorical)
```

```python
for i in categorical:
  temp=dataset.groupby(i)["SalePrice"].count()/len(dataset)
  tempdf=temp[temp>0.01].index
  dataset[i]=np.where(dataset[i].isin(tempdf),dataset[i],"Rare")
```

```python
dataset.head(100)
```

# Encoding Categorical values

```python
for feature in categorical:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    # print(labels_ordered)
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    print(labels_ordered)
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

# Feature Scaling

```python
scaling_feature=[i for i in dataset.columns if i not in ["Id","SalePrice"]]
len(scaling_feature)
```

```python
scaling_feature
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[scaling_feature])
dataset[scaling_feature]=scaler.transform(dataset[scaling_feature])
```

```python
dataset[scaling_feature].head()
```

```python
data=pd.concat([dataset[["Id","SalePrice"]].reset_index(drop=True),dataset[scaling_feature]],axis=1)
```
